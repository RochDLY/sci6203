```python
from pyzotero import zotero
zot = zotero.Zotero("4764734", "group")
items = zot.top()
# we've retrieved the latest five top-level items in our library
# we can print each item's item type and ID
for item in items:
    print('Item: %s | Coord: %s' % (item['data']['title'], item['data']['extra']))
```

```python
for item in items:
   print(item['data']['extra'].split(',')[0], item['data']['extra'].split(',')[1], item['data']['relations'])
```

```python
m = folium.Map([45.4533, -72.0], zoom_start=8)
for item in items:
    
    folium.Marker([item['data']['extra'].split(',')[0], item['data']['extra'].split(',')[1]],
        popup = item['data']['title'],
    ).add_to(m)
m
```

```python
import branca

m = folium.Map([45.4533, -72.0], zoom_start=8)

for item in items:
    html = '<h1>'+item['data']['title']+'</h1>'+'<br/> <b>Légende:</b> '+item['data']['abstractNote']+'<br/> <b>Date:</b> '+item['data']['date']+'<br/> <b>Lieu:</b> '+item['data']['place']

    iframe = branca.element.IFrame(html=html,width=510,height=280)
    popup = folium.Popup(folium.Html(html, script=True), max_width=500)
    
    folium.Marker([item['data']['extra'].split(',')[0], item['data']['extra'].split(',')[1]],
        popup = popup,
        ).add_to(m)

m
```

```python
m.save('carte.html')
```
